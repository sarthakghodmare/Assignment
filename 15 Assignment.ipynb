```python
Q1. Explain GET and POST methods.
soln:
GET Method
Purpose: Request data from a server.

Characteristics:
Idempotent: Same request yields same response.
Safe: Doesn't change server state.
Data in URL: Parameters are in the URL.
Cacheable: Responses can be cached.

POST Method
Purpose: Send data to a server to create/update a resource.

Characteristics:
Non-idempotent: Multiple requests can yield different outcomes.
Data in Body: Parameters are in the request body.
Used for Changes: Alters server state.
Not Cacheable: Responses are not cached.
```

```python
Q2. Why is request used in Flask?
soln:In Flask, the request object is used to access data sent by the client to the server. It encapsulates all incoming request data, providing a convenient way to interact with query parameters, form data, headers, cookies, and more

Accessing Query Parameters: Retrieve data sent in the URL query string.
Accessing Form Data: Retrieve data submitted via HTML forms.
Accessing JSON Data: Retrieve JSON data sent in the request body.
Accessing Headers: Retrieve HTTP headers from the request.
Accessing Cookies: Retrieve cookies sent by the client.
```

```python
Q3. Why is redirect() used in Flask?
soln:In Flask, the redirect() function is used to redirect the client to a different URL. It is commonly used to navigate users to a different page, update the URL, or manage the flow of an application.

After Form Submission:
Redirect users to a different page after a form is successfully submitted to prevent duplicate form submissions if the user refreshes the page.

Handling Authentication:
Redirect users to a login page if they are not authenticated.

URL Management:
Redirect to a different URL or route for URL restructuring or handling different versions of an API.
```

```python
Q4. What are templates in Flask? Why is the render_template() function used?
soln:Templates in Flask are HTML files with placeholders for dynamic content. They allow you to separate the presentation of data from the logic of your application. Flask uses the Jinja2 template engine to render these templates, enabling the insertion of Python variables and control structures into HTML.

The render_template() function is used to render HTML templates with dynamic content in Flask.

Inject Data: Pass data from the server-side Python code into the HTML template.
Display Templates: Render HTML pages with dynamic content based on the data provided.
```

```python
from flask import Flask, render_template, request, jsonify

app = Flask(__name__)

@app.route("/", methods=['GET', 'POST'])
def home_page():
    return render_template('index.html')

@app.route('/mobile', methods=['POST'])
def mobile_operation():
    if request.method == 'POST':
        brand = request.form['brand']
        model = request.form['model']
        price = request.form['price']
        storage = request.form['storage']
        color = request.form['color']

        result = f"Brand: {brand}, Model: {model}, Price: ${price}, Storage: {storage}, Color: {color}"
        
        return render_template('results.html', result=result)

@app.route('/postman_data', methods=['POST'])
def mobile_operation1():
    if request.method == 'POST':
        data = request.get_json()
        brand = data.get('brand')
        model = data.get('model')
        price = data.get('price')
        storage = data.get('storage')
        color = data.get('color')

        result = f"Brand: {brand}, Model: {model}, Price: ${price}, Storage: {storage}, Color: {color}"
        
        return jsonify({'result': result})

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
