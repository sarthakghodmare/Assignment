```python
import seaborn as sns
import plotly.express as px

titanic=sns.load_dataset("titanic")

fig=px.scatter(titanic,x='age',y='age',color='class', title='Scatter Plot of Age vs Fare',
    labels={'age': 'Age', 'fare': 'Fare'},
    hover_data=['sex', 'embarked'] )
fig.show()
```

```python
Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
```

```python
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a box plot
fig = px.box(tips, x="day", y="total_bill", title="Box Plot of Total Bill by Day")

# Show the figure
fig.show()
```

```python
Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.
```

```python
import plotly.express as px

tips=px.data.tips()

fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    pattern_shape="smoker",  # Use the 'smoker' column for pattern shapes
    color="day",             # Use the 'day' column for color
    title="Histogram of Total Bill by Sex with Smoker Pattern and Day Color"
)

# Show the figure
fig.show()
```

```python
Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.
```

```python
import plotly.express as px

iris=px.data.iris()

fig=px.scatter_matrix(iris,
               dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width" ],
               color="species",
               title="Scater matrix for iris")
fig.show()
               
```

```python
Q5. What is Distplot? Using Plotly express, plot a distplot.
```

```python

A Distplot (distribution plot) is a visualization that combines a histogram and a kernel density estimate (KDE) in a single plot. It helps to understand the distribution of a dataset by showing both the frequency of data points (via the histogram) and the underlying probability density (via the KDE).
```

```python
import plotly.express as px

# Load the iris dataset
iris = px.data.iris()

# Create a distribution plot
fig = px.histogram(
    iris, 
    x="sepal_length", 
    nbins=30, 
    histnorm='probability density',  # Normalize to probability density
    title="Distribution Plot of Sepal Length"
)

# Show the figure
fig.show()
```
