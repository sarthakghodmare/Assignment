```python
Q1. What is Flask Framework? What are the advantages of Flask Framework?
soln:
### What is Flask Framework?

**Flask** is a lightweight web framework for Python that provides the essentials for building web applications. It follows the WSGI standard and is known for its simplicity and flexibility.

### Advantages of Flask Framework

1. **Lightweight**: Minimal core with only essential features.
2. **Flexible**: Allows for customizable project structure and components.
3. **Easy to Learn**: Simple API and minimal boilerplate code.
4. **Extensible**: Supports numerous extensions for added functionality.
5. **Good Documentation**: Comprehensive and accessible guides.
6. **Built-in Development Server**: Facilitates local testing and debugging.
7. **Community Support**: Active community and many third-party resources.
```

```python
Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the mapping of URL paths to specific functions in your web application. This is done using the @app.route() decorator, which associates a URL with a view function that processes requests to that URL and returns the appropriate response.

URL Mapping: Links URLs to view functions.
User-Friendly URLs: Creates memorable and navigable URLs.
RESTful Design: Supports HTTP methods for REST APIs.
Modularity: Organizes code into manageable pieces.
Dynamic Routing: Handles URLs with variable parts.
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
```

```python
The url_for() function is used in Flask for URL building. It allows you to dynamically build URLs for your application by referring to the function name that handles the route. This ensures that if you change the URL structure of your application, you only need to update it in one place.

from flask import Flask ,url_for
from flask import request

app = Flask(__name__) 







@app.route('/')
def home():
    return '''
    <h1>Home Page</h1>
    <p>Go to the <a href="{}">welcome page</a>.</p>
    <p>Go to the <a href="{}">contact page</a>.</p>
    '''.format(url_for('welcome'), url_for('contact'))

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/contact')
def contact():
    return '''
    Company Name: ABC Corporation<br>
    Location: India<br>
    Contact Detail: 999-999-9999
    '''





   

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
