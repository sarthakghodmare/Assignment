```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
soln:The subplot() function in Matplotlib is used to create a grid of subplots within a single figure, allowing you to display multiple plots side by side or stacked vertically. This is useful for comparing different datasets or visualizing related data in one cohesive view.
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2= np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10,8))

#subplot1
plt.subplot(2,2,1)
plt.plot(x,y1,marker='o',color='b',label='line1')
plt.title("Line Plot 1")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()

#subplot2
plt.subplot(2,2,2)
plt.plot(x,y1,marker='o',color='r',label='line2')
plt.title("Line Plot 2")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()

#subplot3
plt.subplot(2,2,3)
plt.plot(x,y1,marker='o',color='g',label='line3')
plt.title("Line Plot 3")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()


#subplot4
plt.subplot(2,2,4)
plt.plot(x,y1,marker='o',color='y',label='line4')
plt.title("Line Plot 4")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
A bar plot (or bar chart) is a graphical representation of categorical data using rectangular bars, where the length of each bar corresponds to the value it represents. 
```

```python
Bar plots are useful for comparing different categories and visualizing distributions in a clear and straightforward manner.
```

```python
import numpy as np
import matplotlib.pyplot as plt

#data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


# Create a figure for the plots
plt.figure(figsize=(12,6))

#vertical subplot
plt.subplot(1,2,1)
plt.bar(company,profit,color='skyblue')
plt.title("Vertical Subplot")
plt.xlabel("company")
plt.ylabel("profit")
plt.grid(axis='x')


#horizontal subplot
plt.subplot(1,2,2)
plt.barh(company,profit,color='salmon')
plt.title("Horizonatl Subplot")
plt.xlabel("profit")
plt.ylabel("company")
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
A box plot (or box-and-whisker plot) is a graphical representation of the distribution of a dataset that highlights its central tendency, variability, and potential outliers. It displays the dataset's quartiles and median through a box, while "whiskers" extend to show the range of the data.
```

```python
Summary of Data: Box plots provide a visual summary of key statistical measures, such as the median, quartiles, and range.
Comparison: They are effective for comparing distributions between multiple groups or datasets.
Outlier Detection: Box plots make it easy to identify outliers within the data, which can be useful for further analysis.
```

```python
import numpy as np
import matplotlib.pyplot as plt

#data
box1 = np.random.normal(100, 10, 200) # Normal distribution with mean 100 and std dev 10
box2 = np.random.normal(90, 20, 200) # Normal distribution with mean 90 and std dev 20


plt.figure(figsize=(8,6))
plt.boxplot([box1,box2],labels=['BOX1','BOX2'])
plt.title("BOX PLOT")
plt.xlabel(" X axis")
plt.ylabel("values")
plt.grid(axis='x')
plt.show()
```
