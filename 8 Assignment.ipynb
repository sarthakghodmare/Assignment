```python
class AgeValidationError(Exception):
    def __init__(self, message):
        super().__init__(message)  # Initialize the base Exception class
```

```python
def validate_age(age):
    if age < 0:
        raise AgeValidationError("Entered age is negative")
    elif age > 200:
        raise AgeValidationError("Entered age is very high")
    else:
        print("Age is valid")
```

```python
try:
    age = int(input("Enter your age: "))
    validate_age(age)  # Validate the age
except AgeValidationError as e:
    print(f"Error: {e}")  # Handle the custom exception
```
