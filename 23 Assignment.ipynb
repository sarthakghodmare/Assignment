```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
1.Scatter Plot (scatterplot)

Use: Displays the relationship between two continuous variables. It helps identify correlations, trends, and potential outliers within the data.

2.Box Plot (boxplot)

Use: Visualizes the distribution of a continuous variable across different categories. It highlights the median, quartiles, and potential outliers, making it useful for comparing distributions.

3.Histogram (histplot)

Use: Shows the distribution of a single continuous variable. It helps in understanding the underlying frequency distribution, identifying skewness, and detecting outliers.

4.Bar Plot (barplot)

Use: Displays the mean (or other statistical measure) of a continuous variable across different categories. It is useful for comparing categorical data and visualizing group differences.

5.Heatmap (heatmap)

Use: Visualizes data in a matrix format, often used for correlation matrices or cross-tabulations. It provides insights into patterns, relationships, and the intensity of data points across two dimensions.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

fmri=sns.load_dataset("fmri")
```

```python
fmri
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=fmri,x="timepoint",y="signal",hue="event",style="region",markers=True,dashes=False)
plt.title("LINE PLOT FOR FIMRI DATASET")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.grid()
plt.legend(title='Event and Region', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#data
titanic=sns.load_dataset("titanic")

plt.figure(figsize=(12,6))

#subplot1
plt.subplot(1,2,1) # 1 row, 2 columns, first subplot
sns.boxplot(data=titanic,x ='pclass', y = 'age' )
plt.title('BOX PLOT 1')
plt.xlabel('pclass')
plt.ylabel('Age')

#subplot2
plt.subplot(1,2,2) # 1 row, 2 columns, 2 subplot
sns.boxplot(data=titanic,x ='pclass', y = 'fare' )
plt.title('BOX PLOT 2')
plt.xlabel('pclass')
plt.ylabel('fare')

plt.tight_layout()
plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds=sns.load_dataset("diamonds")

```

```python
diamonds
```

```python
plt.figure(figsize=(12,6))

sns.histplot(data=diamonds,x='price',hue='cut')
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
flights=sns.load_dataset("flights")
```

```python
flights
```

```python
# Pivot the dataset to get it in the right format
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
flights_pivot
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=flights_pivot,annot=True,fmt='g',cmap='YlGnBu', linewidths=0.5)

# Customize the plot
plt.title('Number of Passengers per Month per Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
